```python
def transform_ingredient(vege_):
    a_ = vege_.split("\n")[1:-2]
    b_ = [removal_spaces(i) for i in a_]
    c_ = [i.replace("(",",").replace(")","").split(",") for i in b_]
    vege_list =[[i2.lower().strip() for i2 in i] for i in c_]
    return vege_list
```

```python
def matching_loca(pro_,quality,df_set =[ingre_df2,loca_df]):
    ingre_df,loca_df = df_set
    pro_ = pro_.strip()
    pro_ = pro_.lower()
    ma_ = ingre_df[ingre_df["eng"]==pro_]
    
    try:
        pri_ = ma_["price"].values[0]
        key_ = match_dic[ma_["cate"].values[0]]
        out_ = loca_df[loca_df["cate"]==key_].sample()
    except:
        key_ = random.choice(["fish","agr"])
        out_ = loca_df[loca_df["cate"]==key_].sample()
        pri_ = price = round(np.random.uniform(low=3,high=10,size=(1))[0],2)
        
    out_["ingre"] = pro_
    high_ = min(quality+0.15,1)
    qu_ = np.random.uniform(low=quality,high=high_,size=(1))[0]
    out_["quality"] = qu_
    out_["price"] = round(pri_,2)
    out_["discounted"] = round(pri_*qu_,2)
    return out_
```
